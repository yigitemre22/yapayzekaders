### 4- Feature & Label (Girdi % Sonuç) Ayrımı

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder,OneHotEncoder
label_encoder=LabelEncoder()
all_data["satisfaction"]=label_encoder.fit_transform(all_data["satisfaction"].str.strip())
print(all_data["satisfaction"].unique())
```

```python
columns_cat=["Gender","Customer Type","Type of Travel","Class"]
onehot_coder=OneHotEncoder(handle_unknown="ignore",sparse_output=False)
encoded=onehot_coder.fit_transform(all_data[columns_cat])
encoded_df=pd.DataFrame(encoded,columns=onehot_coder.get_feature_names_out(columns_cat),index=all_data.index)
```

```python
all_data.head()
```

```python
all_data=pd.concat([all_data.drop(columns=columns_cat),encoded_df],axis=1)
all_data.head()
```

```python
X=all_data.drop(["satisfaction"],axis=1)
y=all_data["satisfaction"]
cols=X.columns
cols
```

```python
y.value_counts()
```

### 5- Train - Test Split & Feature Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler

standart_scaler=StandardScaler()
X=standart_scaler.fit_transform(X)
X=pd.DataFrame(X,columns=cols)
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

## Model Geliştirme

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
```

```python
def sklearn_knn(train_data,label_data,test_data,k):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_data,label_data)
    predict_label=knn.predict(test_data)
    return predict_label
    
```

```python
y_predict=sklearn_knn(X_train,y_train,X_test,3)
y_predict
```

```python
y_predict.shape
```

```python
roc_auc=[]
k_value=range(1,15)
for k in k_value:
    y_predict=sklearn_knn(X_train,y_train,X_test,k)
    score=roc_auc_score(y_test,y_predict)
    roc_auc.append(score)

plt.xlabel("k")
plt.ylabel("roc auc score")

plt.plot(k_value,roc_auc,c="g")
plt.grid()
plt.show()
```

En iyi sonuç veren k değeri 7 olduğu için doğruluk değerlerini k=7'ye göre hesaplayacağız.

```python
y_predict_new=sklearn_knn(X_train,y_train,X_test,7)
y_predict_new_train=sklearn_knn(X_train,y_train,X_train,7)

print("modelin roc auc skoru %:",roc_auc_score(y_test,y_predict_new)*100)

print("modelin train set skoru %:",roc_auc_score(y_train,y_predict_new_train)*100)
print("modelin test set skoru %:",roc_auc_score(y_test,y_predict_new)*100)

print("\nclassification report\n",classification_report(y_test,y_predict_new))
print("confusion matrix\n",confusion_matrix(y_test,y_predict_new))
```
