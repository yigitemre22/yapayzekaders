## Model Geliştirme

#### GaussianNB,BernoulliNB,MultinomialNB 

```python
import pygad
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# X: Görüntüdeki normalize edilmiş verilerin olduğu DataFrame
# y: Memnuniyet (Target) sütunu (0 ve 1 şeklinde)

# 1. Uygunluk (Fitness) Fonksiyonu
def fitness_func(ga_instance, solution, solution_idx):
    # 'solution' dizisi 0 ve 1'lerden oluşur [1, 0, 1, 1...]
    selected_features = np.where(solution == 1)[0]
    
    # Eğer algoritma hiç özellik seçmezse cezalandırıyoruz
    if len(selected_features) == 0:
        return 0
    
    # Sadece seçilen sütunları alıyoruz
    X_subset = X.iloc[:, selected_features]
    
    # Model olarak hızlı olduğu için Karar Ağacı kullanıyoruz
    model = DecisionTreeClassifier()
    
    # 3-Katlı çapraz doğrulama ile başarıyı ölçüyoruz
    scores = cross_val_score(model, X_subset, y, cv=3)
    fitness = scores.mean()
    
    return fitness

# 2. GA Parametrelerini Belirleme
num_genes = X.shape[1] # Toplam sütun sayısı

ga_instance = pygad.GA(num_generations=20, # 20 nesil boyunca evrimleş
                       num_parents_mating=5,
                       fitness_func=fitness_func,
                       sol_per_pop=10, # Her nesilde 10 farklı kombinasyon dene
                       num_genes=num_genes,
                       gene_space=[0, 1], # Sadece 0 veya 1 olabilir
                       parent_selection_type="sss", # Stochastic universal sampling
                       crossover_type="single_point",
                       mutation_type="random",
                       mutation_percent_genes=10)

# 3. Algoritmayı Çalıştır
ga_instance.run()

# 4. Sonuçları Görüntüle
solution, solution_fitness, solution_idx = ga_instance.best_solution()
print(f"En iyi başarı oranı: {solution_fitness}")
print(f"Seçilen Sütun İndeksleri: {np.where(solution == 1)[0]}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 1. Sadece GA'nın seçtiği sütunları alıyoruz
X_ga = X.iloc[:, [2, 3, 4, 7, 8, 9, 10, 12, 14, 15, 18, 20, 22, 23, 24, 26]]

# 2. YSA Modelini Oluşturma
model = Sequential([
    Dense(32, activation='relu', input_shape=(16,)), # Giriş ve ilk gizli katman
    Dense(16, activation='relu'),                   # İkinci gizli katman
    Dense(1, activation='sigmoid')                  # Çıkış katmanı (Binary Classification)
])

# 3. Modeli Derleme
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 4. Eğitim
history = model.fit(X_ga, y, epochs=50, batch_size=32, validation_split=0.2)
```
